# Método gráfico de resolución de problemas Lineales


Considere encontrar la solución máxima de la siguiente función objetivo

Z = 4x + 3y

donde X e Y son las variables de decisión.

La función objetivo se encuentra sujeta a las siguientes restricciones:

  x ≥ 0  
  y ≥ 2  
  2y ≤ 25 - x  
  4y ≥ 2x - 8  
  y ≤ 2x - 5  

Se ilustra la implementación en Python de la resolución del problema lineal mediante el método gráfico.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Construct lines
# x > 0
x = np.linspace(0, 20, 2000)
# y >= 2
y1 = (x*0) + 2
# 2y <= 25 - x
y2 = (25-x)/2.0
# 4y >= 2x - 8 
y3 = (2*x-8)/4.0
# y <= 2x - 5 
y4 = 2 * x -5

# Make plot
plt.plot(x, y1, label=r'y1: $y\geq2$')
plt.plot(x, y2, label=r'y2: $2y\leq25-x$')
plt.plot(x, y3, label=r'y3: $4y\geq 2x - 8$')
plt.plot(x, y4, label=r'y4: $y\leq 2x-5$')
plt.xlim((0, 16))
plt.ylim((0, 11))
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')

# Fill feasible region
y5 = np.minimum(y2, y4)
y6 = np.maximum(y1, y3)
plt.fill_between(x, y5, y6, where=y5>y6, color='grey', alpha=0.5)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

La solución se encuentra dentro de la región factible coloreada en gris en el gráfico anterior.

Se ha demostrado que los mínimos y máximos de problemas de programación lineal se encuentran en los vértices de la región factible. En este ejemplo, sólo hay 4 esquinas en la región factible, así que podemos determinar las soluciones para cada esquina y así encontrar el valor que maximiza la función objetivo.

Los cuatro vértices están entre las líneas:

| Línea 1        | Línea 2        | 
| ------------- |:-------------:| 
| y ≥ 2         | 4y ≥ 2x - 8   | 
| 2y ≤ 25 - x   | y ≤ 2x - 5    | 
| 2y ≤ 25 - x   | 4y ≥ 2x - 8   |
| y ≥ 2         | y ≤ 2x - 5    |

Dada la función objetivo,

Z = 4x + 3y

se puede calcular el valor de Z en cada vértice.

Vértice (1)

  y ≥ 2 , 4y ≥ 2x - 8  
  2 = (2x - 8) / 4  
  x = 8  
  y = 2  
  Z = 38  


Vértice (2)

  2y ≤ 25 - x , y ≤ 2x - 5  
  (25 - x) / 2 = (2x - 5)  
  x = 7  
  y = 9  
  Z = 55  


Vértice (3)

  2y ≤ 25 - x , 4y ≥ 2x - 8  
  (25 - x) / 2 = (2x - 8) / 4  
  x = 14.5  
  y = 5.25  
  Z = 73.75  

Vértice (4)

  y ≥ 2 , y ≤ 2x - 5  
  2 = 2x-5  
  x = 3.5  
  y = 2  
  Z = 20  

Por lo tanto el valor máximo para Z es 73.75, cuando X es 14.5 e Y es 5.25.

El método gráfico está limitado a un número de variables y restricciones. Por ejemplo, para la función objetivo:

Z = Ax + By + Cz

habría que graficar en tres dimensiones (x, y y z) lo que complica esta estrategia de análisis.

# Biblioteca Pulp para resolver problemas de optimización
<a href='#contenidos'>Tabla de Contenidos</a>


* Biblioteca PULP: http://pythonhosted.org/PuLP/
* Para instalar en Conda: conda install -c conda-forge pulp 

```python
import pulp
```

PULP es una biblioteca de código abierto de programación lineal. En la presente sección
se resolverá el mismo problema que en la Sección 1 utilizando esta biblioteca.
Recordemos que la función objetivo está dada por

Z = 4x + 3y

sujeta al siguiente conjunto de inecuaciones:

  x ≥ 0  
  y ≥ 2  
  2y ≤ 25 - x  
  4y ≥ 2x - 8  
  y ≤ 2x - 5 
  
Comencemos...

**Paso1:** Inicialicemos una instancia de la clase defina por nuestro problema, como buscamos el óptimo utilizamos el parámetro LpMaximize


```python
miProblema = pulp.LpProblem("Mi primer problema de programacion Lineal", pulp.LpMaximize)
```

En el paso anterior instanciamos nuestra clase (la inicializamos), Ahora a definir las cuáles serán las variables y su rango, la función
objetivo y el conjunto de condiciones a las que estará sujeto el problema

**Paso 2:** Definimos variables y sus límites, para x tiene un límite inferior de 0 y para y un límite inferior de 2

```python
x = pulp.LpVariable('x', lowBound=0, cat='Continuous')
y = pulp.LpVariable('y', lowBound=2, cat='Continuous')
```

**Paso 3:** Definamos la función objetivo

```python
#Función objetivo
miProblema += 4 * x + 3 * y, "Z"
```

**Paso 4:** Agreguemos el conjunto de inecuaciones

```python
#Condiciones de borde del problema
miProblema += 2 * y <= 25 - x
miProblema += 4 * y >= 2 * x - 8
miProblema += y <= 2 * x - 5
```

**Paso 5:** Veamos que hemos construído hasta ahora

```python
miProblema
```

**Paso 6:** Llamamos a la función solve() para resolver el problema. Por defecto Pulp implementa el solver CBC (de COIN-OR Branch and Cut, ver https://projects.coin-or.org/Cbc) utilizado para resolver problemas de complejos de investigaciones de operaciones.
    

```python
miProblema.solve()
```

Veamos el estado del solver, tiene cinco códigos que determinan el estado del cálculo:
* **Not Solved**: Problema sin resolver.
* **Optimal**: Solución óptima encontrada
* **Infeasible**: No hay soluciones factibles (por ej. si se establecen contornos como x <= 1 y x >=2).
* **Unbounded**: Contornos no cerrados, al maximizar la solución tenderá a infinito (por ej. si solo se establece un contorno  como x >= 3).
* **Undefined**: La solución optima puede existir pero el solver no pudo encontrarla con los parámetros establecidos

```python
pulp.LpStatus[miProblema.status]
```

**Paso 7:** Imprimamos la solución y estudie su problema

```python
for variable in miProblema.variables():
    print("{} = {}".format(variable.name, variable.varValue))
```

Voilà, compare con la solución encontrada en la primera parte del cuaderno. Queda evaluar cuanto vale la solución óptima en mi función objetivo

```python
print(pulp.value(miProblema.objective))
```
